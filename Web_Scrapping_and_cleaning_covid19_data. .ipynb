```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import missingno as msno
import re
```

```python
url = 'https://www.worldometers.info/coronavirus/'
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
## Extracting data from website
table = soup.find(class_ = 'table')
```

```python
## Extracting headers from table. 
headers = table.find_all('th')

titles = []

for i in headers:
    title = i.text
    titles.append(title)
    
print(titles)
```

```python
# Add spaces between words using regex
def add_spaces(i):
    words = re.findall('[A-Z][^A-Z]*', i)
    spaced_text = ' '.join(words)
    return spaced_text

# Apply the function to each column name
column_names = [add_spaces(name) for name in titles]

# Display the modified column names
print(column_names)
```

```python
df = pd.DataFrame(columns= column_names)
df.head()
```

```python
## Now, extracting data from the table
rows = table.find_all('tr')

for i in rows[9:240]:
    data = i.find_all('td')
    row = [tr.text for tr in data]
    l = len(df)
    df.loc[l] = row
    
df
```

```python
df.columns
```

```python
## Drop the columns which are not necessory
drop_columns = ['','New Cases','New Deaths','New Recovered','Serious, Critical','Tot  Cases/1 M pop',
                'Deaths/1 M pop','Tests/\n1 M pop\n','Caseevery  X ppl','Deathevery  X ppl',
                'Testevery  X ppl','New  Cases/1 M pop','New  Deaths/1 M pop','Active  Cases/1 M pop']
df = df.drop(drop_columns, axis=1)

df.head()
```

```python
## Rename column
df = df.rename(columns={'Country, Other': 'Country'})
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.info()
```

```python
## Not getting missing data due to spacing in the dataset
# Removing the unnecessory spaces from the dataset
df = df.replace(r'^\s*$', np.nan, regex=True)
df.head(10)
```

```python
mno_count = df.isna().sum()
mno_count
```

```python
# Count the occurrences of 'N/A' in the dataset
na_count = (df == 'N/A').sum()
na_count
```

```python
# Filling missing values and 'N/A' values with zero

df = df.replace('N/A', 0)
df = df.fillna(0)

df
```

```python
df.info()
```

```python
## Now changing the type of some columns
columns_to_convert = ['Total Cases', 'Total Deaths', 'Total Recovered', 'Active Cases', 'Total Tests', 'Population']

# Iterate over each column name and convert to int
for column in columns_to_convert:
    df[column] = df[column].replace(',', '', regex=True).astype(int)

df.info()
```

```python
# Sort the data by  column 'Total Cases' in descending order
df.sort_values('Total Cases', ascending=False, inplace=True)

df.head(10)
```

```python
# Drop rows where the Continent value is 0
df = df[df['Continent'] != 0]
df.info()
```

```python
df.to_excel('COVID DATA_new.xlsx',index=False)
```
